```python
import pickle
from scipy.sparse import coo_matrix
import numpy as np
import matplotlib.pyplot as plt

f = open('data/train_labels_sparse.pickle', 'rb')
train_labels = pickle.load(f)
train_labels_dense = train_labels.todense()
f = open('data/train_features_sparse.pickle', 'rb')
train_features = pickle.load(f)
train_features_dense = train_features.todense()
```

### Number of non-zero features by row

```python
features_count_by_row = np.array((train_features != 0).sum(axis = 1))
print('max # features:\t\t', features_count_by_row.max())
print('min # features:\t\t', features_count_by_row.min())
print('mean # of features:\t', features_count_by_row.mean())
print('median # of features:\t', np.median(features_count_by_row))
plt.hist(features_count_by_row, bins = 'auto')
plt.title('nonzero features per row')
plt.show()
plt.hist(np.log(features_count_by_row), bins = 'auto')
plt.title('log nonzero features per row');
```

### Number of rows that contain non-zero values for each feature

```python
row_count_by_feature = np.array((train_features != 0).sum(axis = 0)).T
print('max # rows:\t ', row_count_by_feature.max())
print('min # rows:\t ', row_count_by_feature.min())
print('mean # of rows:\t ', row_count_by_feature.mean())
print('median # of rows:', np.median(row_count_by_feature))
plt.hist(row_count_by_feature, bins = 'auto')
plt.title('nonzero rows per feature')
plt.show()
plt.hist(np.log(row_count_by_feature), bins = 'auto')
plt.title('log nonzero rows per feature')
plt.show()
```

### Distribution of mean values by feature

```python
mean_features = train_features_dense.mean(axis = 0).T
plt.hist(mean_features, bins = 'auto')
plt.title('feature mean (including zeros)')
plt.show()
plt.hist(np.log(mean_features),bins = 'auto')
plt.title('log feature mean')
plt.show()
```

```python
nonzero_feature_mean = np.true_divide(train_features_dense.sum(0),(train_features_dense!=0).sum(0)).T
plt.hist(nonzero_feature_mean, bins = 'auto')
plt.title('nonzero feature mean')
plt.show()
plt.hist(np.log(nonzero_feature_mean), bins = 'auto')
plt.title('log nonzero feature mean')
plt.show()
```

### Distribution of max values by feature

```python
max_values_features = train_features_dense.max(0).T
plt.hist(max_values_features, bins = 'auto')
plt.title('max value per feature')
plt.show()
plt.hist(np.log(max_values_features), bins = 'auto')
plt.title('log max value per feature')
plt.show()
```
